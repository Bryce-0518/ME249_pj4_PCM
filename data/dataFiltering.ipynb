```python
import DataEng as DE
```

```python
filepath = 'kim_2019_for_TOS_use.csv'
columns = ['survey.id', 'Preference', 'Tout', 'Ta_chair', 'RH_chair', 'chair.status']
keyColumn = 'survey.id'
surveyID = 1
```

```python
df = DE.select_Columns_fromCSV(filepath, columns, keyColumn, surveyID)
# change chiar.staus column data to chiarStatus
df.columns = ['Tout', 'Ta_chair', 'RH_chair', 'survey.id', 'Preference', 'chairStatus']
df
```

```python
import numpy as np
df['Preference'] = np.where((df.Preference == 'No change'), 0, df.Preference)
df['Preference'] = np.where((df.Preference == 'Cooler'), 1, df.Preference)
df['Preference'] = np.where((df.Preference == 'Warmer'), 2, df.Preference)
df['chairStatus'] = np.where((df.chairStatus == 'No'), 0, df.chairStatus)
df['chairStatus'] = np.where((df.chairStatus == 'Yes'), 1, df.chairStatus)
df
```

```python
df.to_csv('dataPCM.csv')
```
